```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sinesum as ss
```

$t\in[-T/2,T/2]$:
  $$f(t) = \begin{cases}
             1, & 0 < t < T/2 \\
             0, & t = 0 \\
             -1, & -T/2 < t < 0
           \end{cases}$$
  The following sum converges to $f(t)$ as $n\to\infty$:
  $$S_n(t) = \frac{4}{\pi}\sum_{k=1}^n \frac{1}{2k-1} \sin\left(\frac{2(2k-1)\pi t}{T}\right)$$

```python
T = 2/np.pi
#function to create proper x axis for various n values
def xvals(n):
    return np.arange(1, n+1)
def plot(x, y, l):
    a = plt.axes()
    if (len(x) > 1):
        a.plot(x, y, color="m", label=l)
    #single point at n=1
    else:
        a.scatter(x, y, color="m", label=l)
    a.set(xlabel="n", ylabel="Summation")
    plt.show()
```

```python
alpha = 0.01
n = 1
ss1 = ss.s(T, alpha*T, n)
x = xvals(n)
plot(x, ss1, plt.title("Alpha = 0.01, n = 1"))
```

```python
alpha = 0.01
n = 3
ss1 = np.array([ss.s(T, alpha*T, b) for b in x])
x = xvals(n)
plot(x, ss1, plt.title("Alpha = 0.01, n = 3"))
```
